Implementation of EWA PDFs following [[2111.02442]](https://arxiv.org/pdf/2111.02442.pdf)

```julia
using Pkg
Pkg.activate("./one_flav_env")

using HCubature
using PythonPlot
using LaTeXStrings

figdir = "./figs/"

rc("font", family="serif", serif="Computer Modern", size=14)
rc("figure", autolayout=true)
rc("axes", labelsize=16, unicode_minus=false)
rc("text", usetex=true)
;
```

```julia
include("./one_flav_env/muon_pdf.jl")
using .MuPDF
```

```julia
xs = 10 .^ range(-3, 0, 1000)

f = figure()
plot(xs, MuPDF.pdf_γ.(xs, (3e3)^2) + MuPDF.pdf_Z.(xs, (3e3)^2), label=L"\gamma / Z")
plot(xs, MuPDF.pdf_WT.(xs, (3e3)^2), color="tab:red", label=L"W_T")
plot(xs, MuPDF.pdf_WL.(xs, (3e3)^2), color="purple", label=L"W_L")

xscale("log")
yscale("log")

xlim(1e-3, 1e0)
ylim(1e-2, 1e3)

xlabel(L"x")
ylabel(L"f_{i/\mu}(x, Q)")
title(L"$\mu$ PDFs for $Q = 3$ TeV")

grid(alpha=0.3)

legend()

plotclose()
# f.savefig(figdir * "mupdf.pdf")
f
```
